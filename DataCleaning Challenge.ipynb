##  Creating a function to convert all heights to centimeters

```python
def height_to_cm(Height):
    if 'cm' in Height:
        return int(Height.replace('cm',''))
    else:
        feet, inches = Height.split('\'')
        inches = inches.replace('"', '')
        return round((int(feet)*30.48)+(int(inches)*2.54))
```


## Applying the 'height_to_cm' function to the 'Height' column to convert heights to centimeters

```python
df.Height=df.Height.apply(height_to_cm)
df.Height.unique()#confirm the changes
```

## Creating a function to convert all weights to kilograms

```python
def weight_to_kg(Weight):
    if 'kg' in Weight:
        return int(Weight.replace('kg',''))
    else:
        return round(int(Weight.replace('lbs','')) * 0.453592)
```

## Applying the 'weight_to_kg' function to the 'Weight' column to convert weights to kilograms

```python
df.Weight=df.Weight.apply(weight_to_kg)
df.Weight.unique()#confirm the changes
```


## Creating a function to convert money values to numerical values (in euros)

```python
def clean_money(Money):
    Money = Money.replace('€','')
    multiplier = 1
    if Money[-1]=='M':
        multiplier = 1000000
        Money = Money[:-1]
    elif Money[-1]=='K':
        multiplier = 1000
        Money = Money[:-1]
    return round(float(Money) * multiplier)
```


## Applying the 'clean_money' function to the 'Value', 'Wage', and 'Release Clause' columns to convert money values to numerical values (in euros)

```python
df[['Value','Wage','Release Clause']]=df[['Value','Wage','Release Clause']].applymap(clean_money)
```

## Create a function to clean the Hits column

```python
def clean_hits(Hits):
    if 'K' in Hits:
        Hits=Hits.replace('K','')
        Hits=int(float(Hits)*1000)
    else:
        Hits=int(Hits)
    
    return Hits
```

## Fill NaN with 0, apply the clean_hits function to the Hits column, and check the correction made

```python
df.Hits=df.Hits.fillna('0').apply(clean_hits)
df.Hits.unique()#check correction made
```

## Create a function to sort the Contract column.
## The Function returns three values: contract_start, contract_end, and contract_type


```python
def contract_sorting(Contract):
    if Contract == 'Free':
        contract_type = 'Free Agent'
        contract_start = None
        contract_end = None
    elif ' On Loan' in Contract:
        contract_type = 'On Loan'
        contract_start= None
        contract_end= Contract.replace(' On Loan','')
    else:
        contract_type = 'On Contract'
        contract_start, contract_end = Contract.split(' ~ ')

    return contract_start, contract_end,contract_type
```

## Use the apply method on contract_sorting and apply method to put it into three series and assign the values to newly created columns 'Contract Start','Contract End', and 'Contract Type'

```python
df[['Contract Start','Contract End', 'Contract Type']]=df.Contract.apply(contract_sorting).apply(pd.Series)
```
